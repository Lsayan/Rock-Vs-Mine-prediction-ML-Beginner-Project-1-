# ROCK VS MINE DETECTION

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.utils import shuffle
```

```python
#Data Collection and Data Processing
dataset = pd.read_csv('Sonar_Data.csv',header=None) #as no header in the data
```

```python
dataset.head()
```

```python
dataset = shuffle(dataset)
```

```python
dataset.head()
```

```python
#No of rows and columns
dataset.shape
```

```python
dataset.describe()
```

```python
#find no of examples
dataset[60].value_counts()
```

```python
# M for mine
# R for rock
```

```python
dataset.groupby(60).mean()
```

```python
#for seperating the data  and labels
X = dataset.drop(columns = 60,axis=1)
Y = dataset[60]
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify = Y,random_state=1)  #stratify based on proportional data set
```

```python
X_train.shape, X_test.shape
```

```python
#We will be using logistic regression
```

```python
Model = LogisticRegression()
```

```python
#Training the Logistic Regression using the training Data
```

```python
Model.fit(X_train,Y_train)
```

```python
#Model has been trained
```

```python
#Accuracy for the training Data
```

```python
X_train_pred = Model.predict(X_train)
#It gives the accuracy score
training_data_accuracy = accuracy_score(X_train_pred,Y_train)
```

```python
print(f'The accuracy of the training dataset is: {training_data_accuracy}')
```

```python
X_test_pred = Model.predict(X_test)
```

```python
#For the testset data
test_data_accuracy = accuracy_score(X_test_pred,Y_test)
```

```python
print(f'The accuracy of the test data set is: {test_data_accuracy}')
```
